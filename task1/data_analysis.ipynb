```python
import pandas as pd
import plotly.express as px
```

```python
df_train = pd.read_csv("processed/train.csv")
```

```python
fig = px.histogram(df_train, x='rating', width=800, title="Rating Distribution")
fig.update_xaxes(tickmode="linear")
fig.update_traces(xbins_size=0.5)
```

```python
fig = px.histogram(df_train, x='part', width=800, title="Positive/Negative Distribution")
fig.update_xaxes(tickmode="linear")
fig.update_traces(xbins_size=0.1)
```
